```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(erosion),plt.title('Averaging')
```

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(opening),plt.title('Averaging')
```

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(closing),plt.title('Averaging')
```

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(gradient),plt.title('Averaging')
```
