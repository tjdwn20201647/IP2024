#### Rotate

```python
img = cv2.imread('cutedog.png',0)
rows,cols = img.shape
M = cv2.getRotationMatrix2D((cols/5,rows/5),90,1) #값의 의미 이해
dst = cv2.warpAffine(img,M,(cols,rows))

cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Afine Transformation

```python
img = cv2.imread('cutedog.png')
pts1 = np.float32([[50,50],[200,50],[50,200]]) #점 3개로 이동, 유지하며 이동
rows,cols,ch = img.shape
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

#### Perspective Transformation

```python
img = cv2.imread('IMG_6504.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[170,190],[760,195],[808,1061],[100,1080]])
pts2 = np.float32([[0,0],[400,0],[400,600],[0,600]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(400,600))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('dsu5.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[554,231],[690,273],[699,359],[558,334]])
pts2 = np.float32([[0,0],[500,0],[500,250],[0,250]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(500,250))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
