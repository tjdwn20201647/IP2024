```python
import numpy as np
import cv2
```

```python
img = cv2.imread('dog2.jpg',0)
```

```python
cv2.imshow('image',img) #'image' 창의 이름
cv2.waitKey(0) #키 입력을 할 때까지 무한 대기
cv2.destroyAllWindows() #모든 OpenCV 윈도우 창 닫기
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL) #창 크기를 자유롭게 조절
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('messigray.png',img) #messigray 이름으로 파일 저장 
```

```python

img = cv2.imread('dog2.jpg',0)
cv2.imshow('image',img)
k = cv2.waitKey(0)
if k == 27: # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('cutedog.png',img)
cv2.destroyAllWindows()
```
